```python
Inferential - Result: based on sample data(probability based)
------------
 |-> take sample data from population - sample data ->analysis ->visualization
                                        -----------
 100's  - 5s ---->5s ... 5s -OK
Includes
 - comparision 
 - estimation
 - relationship
 - predictions
 - testing
results:  probability score 
measures of spread  # how the data will spread 
------------------- 
   |-> range = max - min
   |-> Interquartile Range(IQR) - 1st quartile(Q1) and 3rd quartile(Q3) 
          Q1 - The median of the lower half of the data.
               The value below which 25% of the data
          Q2 - secondard quartile - median of the dataset 50%. it divides the data into two halves
          Q3 - third quartiles - median of the upper half of the data 75%.
                                 It is value below which 75% of data.
         IQR = Q3-Q1

Lowerbound outliers = Q1 - 1.5 * IQR
upperbound outliers = Q3 + 1.5 * IQR

Steps to calculate IQR
------------------------
step 1: sort the data in ascending order 
step 2: Calculate Q1 - median of the lower half of the data
step 3: Calculate Q3 - median of the upper half of the data
step 4: Compute IQR = Q3 - Q1

   |->variance  
          |-> population variance  |-> sample variance 
   |->standard deviation
   |->log -> log5 log10  -> log5-log125 (starts = 5 ; ends = 125)
   |->mean absolute deviation 
```

```python
zscore
--------
     x - (mean of the data point)
z = ------------------------------------------
        standard deviation of the dataset

x = individual data point

z = 0 - data point is exactly at the mean
z = +Ve - data point is above the mean
z = -Ve - data point is below the mean
z = above 2 =>outlier
          |________________threshold 
```

```python
d = np.array([50,60,70,80,90])
mean = np.mean(d)
std_dev = np.std(d)
zscore = (d - mean) / std_dev
zscore
```

```python
outliers = d[np.abs(zscore) > 2 ]
print(outliers)
```
